```python
from rdkit import Chem
import pandas as pd
```

```python
raw = pd.read_csv("intermediate/library+shulgin-raw.csv", converters={i: str for i in range(7)})
```

```python
raw
```

```python
def can(smiles):
    return Chem.MolToSmiles(Chem.MolFromSmiles(smiles),True)
```

```python
raw['can'] = raw.apply(lambda row: can(row.smiles), axis=1)
```

```python
raw
```

```python
# How rows are unchanged?
raw[raw['can'] == raw['smiles']].shape
```

```python
# How many rows have changed?
raw[raw['can'] != raw['smiles']].shape
```

```python
raw.to_csv("intermediate/library+shulgin-can.csv", index=False)
```
